```python
#for i in name:
#    print(name[i])
soup = []
for i in Celeb_Name:
    r1 = rq.get(Celeb_Name[i])
    soup.append(BeautifulSoup(r1.text, "html.parser"))
    
    
```

```python
# define function to save link 
def gen_link(text):
    link = []
    x = text.select('img[src^="https://media.gettyimages.com/"]')
    for img in x:
        link.append(img['src'])
    return link
```

```python
links = []
for i in soup:
    links.append(gen_link(i))
    
```

```python
len(links)
```

```python
links[0]
```

```python
# create folder and download & save images
os.mkdir('dataset')
path = os.getcwd()
os.chdir('./dataset')

ind = 0
for i in Celeb_Name:
    os.mkdir(i)
    j=1
    for index, img_link in enumerate(links[ind]):
        if j<=60:
            img_data = rq.get(img_link).content
            with open(i+'/'+str(index+1)+'.jpg','wb+') as f:
                f.write(img_data)
            j+= 1
        else:
            f.close()
            break
    ind += 1
os.chdir(path)
```
